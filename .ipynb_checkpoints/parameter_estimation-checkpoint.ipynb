```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
```

```python
def synthetic_cmb(x):
    return np.exp(-x**2 / 0.01) * np.cos(50 * x)

def synthetic_bao(x):
    return np.sin(2 * np.pi * x) / (x + 0.1)

x_vals = np.linspace(0, 1, 500)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 4))

# CMB Plot
plt.subplot(1, 2, 1)
plt.plot(x_vals, synthetic_cmb(x_vals), color='midnightblue')
plt.title("Synthetic CMB")
plt.xlabel("Angular Position")
plt.ylabel("Amplitude")

# BAO Plot
plt.subplot(1, 2, 2)
plt.plot(x_vals, synthetic_bao(x_vals), color='darkgreen')
plt.title("Synthetic BAO")
plt.xlabel("Angular Position")
plt.ylabel("Amplitude")

plt.tight_layout()
plt.show()
```

```python
from scipy.signal import find_peaks

# Find peaks in synthetic signals
cmb_peaks, _ = find_peaks(synthetic_cmb(x_vals), height=0.05)
bao_peaks, _ = find_peaks(synthetic_bao(x_vals), height=0.05)

# Angular positions of peaks
cmb_angles = x_vals[cmb_peaks]
bao_angles = x_vals[bao_peaks]

print("CMB Peak Angles:", np.round(cmb_angles, 4))
print("BAO Peak Angles:", np.round(bao_angles, 4))
```

```python
angular_delta = bao_angles[0] - cmb_angles[0]
ratio = bao_angles[0] / cmb_angles[0]

print("Angular Spacing (BAO - CMB):", round(angular_delta, 4))
print("BAO / CMB Ratio:", round(ratio, 4))
```
